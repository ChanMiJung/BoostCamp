# Intro to NLP, Bag-of-Words

## 자연어 처리

### Natural language processing(NLP)
* 자연어처리
* 자연어를 다루는 다양한 분야 중 하나
* 딥러닝을 선도함
* 2 종류의 task로 분류됨
  * Natural Language Understanding(NLU)
    * 컴퓨터가 주어진 단어, 문장, 문단, 글을 이해하는  과정
  * Natural Language Generation(NLG)
    * 자연어를 상황에 따라 적절하게 생성할 수 있는 task 종류
* NLP 연구가 발표되는 곳
  * ACL, EMNLP, NAACL
* NLP가 다루는 task
  * Low-level parsing
    * 각 단어를 준비하기 위한 low-level tasking
    * 문장을 이루는 각 단어(token)들을 정보단위로 생각함
    * Tokenization : 주어진 문장을 단어 단위로 쪼개는 과정
      * 문장은 token들이 순서대로 이루어진 sequence
    * Stemming : 단어의 다양한 의미 변화를 없애고, 그 의미만을 보존하는 단어의 어근을 추출하는 과정
      * ex) 맑다 -> 맑고, 맑은, 맑게
  * Word and phrase level
    * Named entity recognition(NER) : 단일 단어 또는 여러 단어로 이루어진 고유명사를 인식하는 task
      * ex) NewYork Times
    * Part-of-speech(POS) tagging : 단어들이 문장내에서 품사나 성분이 무엇인지 알아내는 task
  * Sentence level
    * Sentiment analysis : 주어진 문장에 대한 감정분석
      * ex) 긍정적인 문장 / 부정적인 문장
    * Machine translation
      * 번역
      * 주어진 문장을 번역하는 경우 주어진 문장을 전체적으로 이해하고, 번역할 때 각 단어별로 적절한 단어로의 번역과 번역되는 언어의 문법에 필요한 어순을 잘 고려하여 수행해야 함
  * Multi-sentence and paragraph level
    * Entailment prediction : 두 문장간의 논리적인 내포 혹은 모순 관계를 예측함
    * question answering : 독해기반의 질의응답
      * 질문에 있는 keyword 들이 포함된 문서들을 먼저 검색한 후, 문서로부터 독해를 통해 질문에 대한 정답을 정확하게 알아내서 사용자에게 직접적으로 제시함
    * Dialog System : 챗봇과 같이 대화를 수행할 수 있는 자연어 처리 기술
    * Summarization(or Document Summarization) : 주어진 문서를 자동으로 한 줄 요약하는 task


### Text mining
* 자연어를 다루는 다양한 분야 중 하나
* 빅데이터 분석과 관련되는 경우가 많음
* Text mining 연구가 발표되는 곳
  * KDD, The WebConf(formerly, WWW), WSDM, CIKM, ICWSM
* 여러 데이터에서 특정 keyword들의 빈도수를 시간 순으로 trend를 분석하고, 그 결과로 특정 인물의 이미지 변화 또는 상품에 대한 소비자 반응을 분석할 수 있음
* Topic modeling(Document clustering)
  * 서로다른 keyword이지만 비슷한 의미를 가지는 keyword들을 grouping 해서 분석하는 과정을 자동으로 수행해주는 기법
  * ex) 가성비, 내구성, 에프터 서비스 등의 세부적인 요소들을 이야기하고 그에대한 정보를 자동으로 빠르게 얻어낼 수 있음
* Computational Social Science와 밀접한 관련이 있음
  * ex) 현대인의 행동 패턴 분석(혼밥 등등), 신조어 등등



### Information retrieval
* 자연어를 다루는 다양한 분야 중 하나
* Information retrieval 연구가 발표되는 곳
  * SIGIR, WSDM, CIKM, RecSys
* Google, Naver 등에서 사용되는 검색 기술을 연구하는 분야
* 검색 system의 검색 성능이 고도화 되면서 검색 기술은 어느정도 성숙한 상태에 이름 -> 상대적으로 기술발전이 느림
* 정보 검색 분야에 포함되는 세부분야 추천 system : 자동화되고 적극적인 새로운 형태의 검색 system

### Trends of NLP
* CV에 비에서 느리지만 꾸준히 발전하고 있음
* Word Embedding : 단어를 vector 공간의 한 점으로 나타냄
  * 단어를 특정 dimension을 가진 vector로 표현함
* 문장을 word vector들이 특정한 순서로 주어지는 sequence라고 할 수 있음
  * Sequence data를 처리하는데 특화된 모델 : RNN(LSTM, GRU)
* 2017년 'Attention is All You Need' 논문을 통해 Transformer(RNN기반의 자연어 처리 구조를 self attention이라고 불리는 module로 완전히 대체할 수 있는 모델)을 가져옴
  * 큰 성능 향상을 가져옴
  * 기계번역 task를 위해 처음으로 제안됨
* 현재 대부분의 자연어처리 딥러닝 모델들은 transformer 모델구조를 기반으로 하고 있음
* 자연어 처리의 자가지도학습
  * 입력 문장이 주어졌을 때, 일부 단어를 가려놓고 맞추도록 하는 task(앞뒤 문맥을 파악하여 단어를 맞춤)
* 자가지도학습을 통한 pretrained model의 예시
  * BERT, GPT-3 : 범용 인공지능 기술

## Bag-of-Words
* NLP와 Text mining에서 딥러닝 기술이 적용되기 이전에 많이 활용되던 기법
* 단어 및 문서를 숫자형태로 나타내는 가장 간단한 기법

### Bag-of-Words Representation
* Step 1. Constructing the vocabulary containing unique words.
  * 주어진 text dataset에서 unique한 word들을 모아서 vocabulary를 구축함(vocabulary dictionary에 단어를 삽입)
    * 중복되는 단어는 1번만 등록

* Step 2. Encoding unique words to one-hot vectors
  * categorical variable : vocabulary의 단어
  * categorical variable로 이루어진 data에 다양한 머신러닝/딥러닝 모델을 적용하기 위해 categorical variable을 one-hot vector로 변환함
    * vocabulary 전체 수를 dimension의 수로 하는 vector공간을 설정(vocabulary dict length의 list 생성)
    * 공간에 존재하는 dimension을 vocabulary에 등록된 각 단어에 mapping하여 해당 dimension의 값만 1, 나머지는 0으로 할당하여 vector를 표현함(vocabulary에서 단어의 index에 해당하는 값을 1, 나머지는 0)
  * 어떤 단어쌍이든 유클리드 distance가 $\sqrt{2}$, cosine similarity(내적, 유사도)은 0
    * 단어의 의미와 상관없이 동일한 관계를 가지는 형태로 단어의 벡터 표현형을 설정함


* Bag-of-Words Vector
  * one-hot vector의 확장
  * 문장 또는 문서에 포함되는 단어들의 one-hot vector를 모두 더하여 문장 또는 문서를 나타냄

* ex)
  * Sentence
    1. "John really really loves this movie"
    2. "Jane really likes this song"
  * Vocabulary : {"John", "really", "loves", "this", "movie", "Jane", "likes", "song"}
  * one-hot vector
    * "John" : [1 0 0 0 0 0 0 0]
    * "really" : [0 1 0 0 0 0 0 0]
    * "loves" : [0 0 1 0 0 0 0 0]
    * "this" : [0 0 0 1 0 0 0 0]
    * "movie" : [0 0 0 0 1 0 0 0]
    * "Jane" : [0 0 0 0 0 1 0 0]
    * "likes" : [0 0 0 0 0 0 1 0]
    * "song" : [0 0 0 0 0 0 0 1]
  * Bag-of-Words Vector
    * "John really really loves this movie" : [1 2 1 1 1 0 0 0]
    * "Jane really likes this song" : [0 1 0 1 0 1 1 1]



### NavieBayes Classifier for Document Classification
* NaiveBayes Classifier : Bag-of-Words를 활용한 대표적인 문서 분류 기법
  * Bag-of-Words Vector로 나타낸 문서를 정해진 category 또는 class 중에 하나로 분류하는 기법

* Bayes' Rule Applied to Documents and Classes
  * 가장 높은 확률분포를 가지는 class `c`를 선택해서 조건부 확률을 계산
  * $c$ : category class, $C$ : total number of category classes
  * $d$ : document
  * $P(c \mid d)$ : 특정 문서 $d$가 주어져 있을 때, 그 문서가 특정 class $c$에 속할 확률
  * $P(d)$ : 특정 문서 $d$가 뽑힐 확률, 여기서는 고정된 하나의 문서이기 때문에 **상수**와 같음, $\mathrm{argmax}$에서 무시 가능
  * $P(d \mid c)$ : $c$가 고정되었을 때 $d$가 나타날 확률
    * $P(d \mid c) = P(w_1, w_2, \cdots, w_n \mid c)$
      * $n$개의 word가 동시에 일어나는 동시사건

* $P(d \mid c)P(c) = P(w_1, w_2, \cdots, w_n \mid c)P(c) \rightarrow P(c)\prod_{w_i \in W}P(w_i \mid c)$
  * 각 단어가 등장할 확률 : $c$가 고정되어 있는 경우 word가 서로 독립이라고 가정하면, 각 단어가 나타날 확률을 모두 곱한 형태로 표현 가능

* Example
  * $P(c)$
    IntroNLP_NavieBayes_Classifier_example.PNG
    * $P(c_{CV}) = {2 \over 4} = {1 \over 2}$
      * 4개의 documents 중에서 cv class에 해당하는 2 개의 documents
    * $P(c_{NLP}) = {2 \over 4} = {1 \over 2}$
      * 4개의 documents 중에서 cv class에 해당하는 2 개의 documents
  * $P(w \mid c)$
    IntroNLP_NavieBayes_Classifier_example2.PNG
    * $P(w_k \mid c_i) = {{n_k} \over n}$
      * $n_k$ : $c_i$ class에서 $w_k$가 발생하는 횟수
