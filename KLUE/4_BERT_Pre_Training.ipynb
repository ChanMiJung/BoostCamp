# 한국어 언어 모델 학습 및 다중 과제 튜닝

## 1. BERT 학습하기

### 1. BERT 모델 학습 단계

* BERT 학습의 단계
  1. Tokenizer 만들기
  2. 데이터셋 확보
    * BERT는 model size가 크기 때문에 항상 대용량의 corpus를 필요로 함
  3. Next sentence prediction (NSP)
  4. Masking

* 새로 학습하는 이유
  * 도메인 특화 task의 겨우, 도메인에 특화된 학습데이터만 사용하는 것이 성능이 더 좋기 때문
    * 새로 학습하는 것이 기존의 BERT를 finetuning하는 것 보다 성능이 좋음

* 모델 학습 기준
  * 데이터셋
    * model에 입력으로 들어갈 수 있는 형태

    * model에 데이터를 입력으로 주는 방법
