## 1. 두 문장 관계 분류 task

* 주어진 2개의 문장에 대해, 두 문장의 자연어 추론과 의미론적인 유사성을 측정하는 task
  * [CLS] token에 classification layer를 부착하여 두 문장의 관계를 분류함

### 2. 두 문장 관계 분류를 의한 데이터

* Natural Language Inference(NLI)
  * 언어모델이 자연어의 맥락을 이해할 수 있는지 검증하는 task
  * 전제문장(Premise)과 가설문장(Hypothesis)을 Entailment(함의), Contradiction(모순), Neutral(중립)으로 분류
  * 분류의 대상
    * 가설문장이 참인 경우
    * 가설 문장이 거짓인 경우
    * 가설 문장이 참일 것으로 추정되거나 명백하게 참, 거짓을 알 수 없는경우
    * 3개의 label을 classification해야함

## 2. 두 문장 관계 분류 모델 학습
* BERT를 활용해 IRQA 챗봇 구현

### 1. Information Retrieval Question and Answering(IRQA)

* 사전에 정의해놓은 QA set에서 가장 적절한 답변을 찾아내는 과정
  * [CLS] token을 이용해서 cosine similarity로 chatbot을 구현한 것과 유사한 task

* 시스템 구조도
  * 사용자 질문
  * BERT를 통해 sentence embedding을 함
  * 사전에 정의한 QA pair table도 sentence embedding함
  * 기존에 존재하던 dataset의 question과 입력된 질문 query의 유사도를 비교하여 가장 적절한 문장을 반환함

* 기존 chatbot과 차이점
  * task 뒷단에 paraphrase detection(두 문장 관계 분류 task를 학습한 모델)을 부착

* paraphrase detection을 부착하는 이유
  * Top-N 개로 문장을 반환할 수 있지만, 그 중에 Top-1이 정답이 되지 않을 수 있음
  * paraphrase detection을 통해 입력 질문 query와 사전에 정의된 question이 실제로 유사한 의미를 가지고 있는지 검증함
  * 검증 필터링을 통과해야 답변이 출력됨
