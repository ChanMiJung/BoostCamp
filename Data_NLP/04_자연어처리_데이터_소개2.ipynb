## 2. 기계 번역 Machine Translation

* WMT(Workshop Machine Translation) 데이터셋
  * 2014년부터 시행된 기계 번역 학회에서 공개한 데이터셋 다국어 번역 데이터이며, 두 언어간의 병렬 말뭉치(두 언어를 pairing 해서 만든 말뭉치)로 구성됨
  * 뉴스, 바이오, 멀티 모달 데이터 등이 제공됨

image.png

* WMT 데이터셋 - 데이터 형식 2020
  * 데이터를 제작할 때 원시 corpus가 존재하고 원시 corpus 에 대한 주석 corpus가 존재하는 경우, 어떤 원시 corpus 에 달려있는지 id로 mapping하기 때문에 **id**가 매우 중요함

image.png

* WMT 데이터셋 - 데이터 구축 2018년 기준
  * 평가용 데이터 : 1,500 개의 영어 문장을 다른 언어로 번역 + 1,500 개의 문장은 다른 언어에서 영어 문장으로 번역
    * 번역은 source language와 target language가 중요함(방향이 중요함)
  * 훈련용 데이터 : 기존에 존재하는 병렬 말뭉치와 단일 언어 말뭉치를 제공

## 3. 요약 Text Summarization

* 2가지 기계 요약
  * 추출 요약(Extract Summerization) : 나열되어 있는 텍스트에서 텍스트를 대표할 것 같은 문장을 자정한 만큼 선택
  * 추상 요약(Abstract Summerization) : 문장을 생성함. 기계 번역이나 발화 생성 task와 같이 language generation(문장을 generation)하는 task

* CNN / Daily Mail
  * 추상 요약 말뭉치
  * 기사에 대하여 사람이 직접 작성한 요약문이 짝을 이루고 있음
  * 학습 데이터 286,817 쌍, 검증 데이터 13,368 쌍, 평가 데이터 11,487 쌍으로 구성
  * 저작권 문제로 URL list 제공함
  * https://github.com/abisee/cnn-dailymail
  * ground-truth : 사람이 만든 정답셋
    * `@entity` : 사람이름이 들어가는 경우 개인정보를 유출하는 것이기 때문에 익명화를 함
  * 국어원이나 AI hub에서 한국어로 된 추상 + 추출 요약 말뭉치를 제공함

## 4. 대화 Dialogue

* DSTC - Dialog System Technology Challenges
  * DSTC1
    * human-computer dialogs in the bus timetable domain
  * DSTC2 and DSTC3
    * human-computer dialogs in the restaurant information domain
  * DSTC4 and DSTC5
    * DSTC4 human-human dialogs in the tourist information domain
  * DSTC6 이후
    * End-to-End Goal Oriented Dialog Learning, End-to-End Conversation Modeling, and Dialogue Breakdown Detection로 확장

image.png

* chatbot, dialog 시스템은 두 가지로 나뉨
  * Open domain dialog(ODD) : 모든 대화를 함
  * task oriented dialog(TOD) : 목적 지향형의 대화를 함

* DSTC1 - 데이터 형식

image.png

* Wizard-of-Oz
  * WoZ 방식으로 수집된 데이터셋이며 대화 상태 추적 데이터와 유사한 형태로 이루어짐
  * WoZ 방식은 대화 수집 방식의 하나로, 참여자가 대화 시스템을 통해 대화를 하고 있다고 생각하게 한 뒤 실제로는 실제 사람이 참여자의 발화에 맞추어 응답을 제시하고 대화를 이끌어가면서 대화를 수집하는 방식

  * https://huggingface.co/datasets/woz_dialogue
