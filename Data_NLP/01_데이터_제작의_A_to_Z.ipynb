## 데이터 수집-가공 설계

### 원시 데이터 수집 방식

* 전산화, 스크래핑, 작업자 작성, 모델 생성 : 적합한 데이터란 무엇인지 기준 세우기
  * 전산화 : 이미 존재하는 텍스트(출간된 책 등등)의 파일이 없다면 직접 타이핑하여 제작하는 것
  * 작업자 작성 : 작업자들을 통해 대화의 문장을 작성하거나 발화를 작성하게 하는 것
  * 모델 생성 : 모델을 통해서 모델이 데이터를 자동 생성하는 것

image.png

### 작업자 선정

* 주석 작업의 난이도와 구축 규모에 맞는 작업자 선정 및 작업 관리
  * 전문가 : 데이터의 품질이 높아질 수 있음. 데이터 양이 적음. 단가가 높음
  * 크라우드 소싱 : 일반 대중을 통해 대량으로 데이터 수집. 단순하고 직관적인 데이터 수집

image.png

### 구축 및 검수 방법 설계

* 구축 작업의 난이도와 구축 규모, 테스크 특성에 맞는 구축 및 검수 방식(전문가, IAA) 설계
  * 파일럿 : 본격적인 구축에 앞서서 소규모로 미리 실험해보는 것. 최소 100개, 많으면 1,000개 또는 2,000개 정도의 작업을 해보는 것을 추천함
  * 본 구축 : 몇 차례에 나누어서 할지, 하나의 데이터를 한 사람만 볼지 여러명이 볼지 정해야함
  * 일반적으로 ML/DL 을 할 때에 만개 단위로 진행하기 때문에 그 중에 10% 정도를 가지고 파일럿을 구축
  * 가이드라인 작성할 때는 최소 100개 정도의 샘플을 직접 해보면서 작업하는 것이 좋음

image.png

* 설계 시 발견하지 못한 이슈들이 파일럿 과정에서 많이 나옴. 이 과정에서 해결하지 못하면 본 구축까지 전파가 되어서 문제를 일으키게 되기 때문에 파일럿 단계에서 해결해야 함
* 고정적인 작업자가 작업을 하는 경우, 파일럿 테스트를 통해 작업자를 선정함

* 본 구축을 하는 과정에서 끊임없이 중간 검수를 하면서 데이터의 품질을 관리하는 것이 중요하기 때문에 본 구축을 차수별로 나누어서(ex. 1차, 2차) 진행하는 것이 일반적임

### 데이터 검수 및 분석

* 평가 지표 설정
  * 진행하기 전에 먼저 평가 지표를 설정해야 함

image.png

* 가이드라인 적합도 분석 : 가이드라인을 얼마나 잘 지키면서 데이터가 만들어졌는지를 사람이 평가함
* 전문가 : 실제 해당 프로젝트를 설계한 사람 또는 의뢰한 레이블 에이젼시의 담당 PM, 숙련된 작업자 등등

* 자동 평가 및 분석 : 기계적인 방법을 이용함
* 일괄 수정 사항 반영 : 가이드라인에 오류가 있거나, 가이드라인 숙지를 잘못하여 특정한 작업자가 동일한 실수를 반복한 것 등을 발견한 경우 이를 일괄적으로 수정하는 프로세스 단계를 생각해야함

### 자연어처리(NLP, natural language processing) 란?

* 인공지능의 한분야, 사람의 언어를 컴퓨터가 알아듣도록 처리하는 인터페이스 역할
* 자연어 이해(NLU, natural language understanding)와 자연어 생성(NLG, natural language generation)으로 구성

* 자연어 처리의 최종 목표:
  * 컴퓨터가 사람의 언어를 이해하고 여러 가지 문제를 수행할 수 있도록 하는 것

### 데이터 분류 방식

* 원천 데이터 장르(도메인) : 문어(기사, 도서 등), 구어(대화 등), 웹(메신저 대화, 게시판 등)
  * 구어 : 음성 기준 transcription(귀로 듣고 사람이 타이핑 하거나 tool을 이용하여 텍스트로 변환)
  * 웹 : 문어성을 띌 수도 있고 구어성을 띌 수도 있음

* 과제의 유형 : 
  * 자연어 이해(형태 분석, 구문 분석, 문장 유사도 평가 등)
  * 자연어 생성(기계 번역, 추상 요약 등)
  * 혼합(챗봇 등)

\+ 자연어처리 데이터를 만들 때는 복잡한 과제도 단순화하여 단계별로 구축
