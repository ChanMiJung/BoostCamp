* 과제의 목적
  * 관계 추출(Reation Extraction)이란 문장에 등장하는 두 개체 간의 관계를 주석하는 것
  * 관계 추출의 대상이 되는 개체명을 인식하고, 각 개체가 주체(subject)인지 대상(object)를 파악한 뒤 그 둘 간의 관계를 주석함
  * 주체와 대상, 관계로 이루어진 트리플(triplet)을 완성함

* 특정한 도메인에 맞추어 관계 분류(Class) 목록을 확장하고, 확장된 관계를 주석한 데이터를 만듦

* 데이터 평가
  * 작업자간 일치도(IAA, Inter-Annotator Agreement):
    * Fleiss' 𝛋 (TACRED) : 작업자가 3명 이상인 경우 사용함
      * Annotator를 고정해놓고 sampling을 해서 해당 Annotator에 대한 𝛋 value를 살펴보는 것
      * 결측치가 있는 경우에 사용하기 어려움
      * Annotator 중에 일부거 Annotating에 참여하지 않으면 데이터를 사용할 수 없음
    * Krippendorff's α(KLUE)
      * 결측치가 있어도 보정해서 사용할 수 있음
      * 0.70 이상이면 신뢰할만 함


  * 모델 성능 평가:
    * 정밀도(Precision), 재현율(Recall), F1 (TACRED)
    * Micro F1(label 마다 분포가 불균형한 경우 평가하는데 더 유리함), AUPRC(area under the precision-recall curve) (KLUE)
