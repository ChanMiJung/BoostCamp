## 2. 가이드라인의 구성 요소

* 가이드라인을 잘 작성하는 경우 데이터의 품질 향상으로 이어짐
* 작업 의도, 이유를 잘 설명해주는 경우 목적에 맞는 데이터를 조금 더 효율적으로 수집할 수 있음
* 데이터 양에 따라 급여가 바뀔 수 있어서 반려 및 통과 기준을 잘 알고 있어야 함

## 3. 가이드라인의 버전 관리

* 가이드라인은 구축과 검수 과정을 통해 지속적으로 개정되어야 함
* 개정 전과 개정 후 어떤 것이 변화하였는지 비교하여 볼 수 있도록 버전 관리가 되는 것이 가장 좋음
* 버전별로 작성자와 개정일을 기입해두는 것이 관리에 도움이 됨

## 5. 가이드라인 작성 시 유의 사항

* 가이드라인의 유형별 특성을 이해하고 그에 알맞는 정보를 작성함
* 작업자의 작업 이해도를 고려하여 작성함
* 작업자에게 공개해야 하는 필수 정보와 부가적인 정보가 무엇인지 사전에 고려함
* 가이드라인 구성 요소의 배치를 어떻게 할 지 고민함
* 작업자의 가독성을 고려함
